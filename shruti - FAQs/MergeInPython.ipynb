# Merge operations in Python

```python
import pandas as pd

#read csv files in dataframe
df_rain = pd.read_csv('rainfall201701.csv')
df_site = pd.read_csv('sites201701.csv')
```

```python
df_rain.head()
```

```python
df_site.head()
```

# Merging dataframes

Let's say we want to see the longitude and longitude information, with rainfall information jointly for corresponding id and name.

```python
df_merged = pd.merge(df_rain,df_site,how='inner',on=['id','name'])
df_merged
```

According to your analysis, you can specify join operation to be how= 'outer', how= 'left', how= 'right' in above line of code. Also the parameter "on = " takes the key values on which the merge operation is performed.
